## Diccionario de Datos


Para mejorar nuestra comprensión de los datos y la informacion que nos transmite creamos un diccionario de los atributos, con esto podremos evaluar las relaciones y derivar el analisis que realizaremos.


1. work_year = Años de ingreso al trabajo 
2. experience_level = nivel de experiencia (EN Entry level - Junior, MI Middle level, SE Senior level, EX Executive level) 
3. employment_type = tipo de empleo (FullTime, PartTime, Contract, Freelancer) 
4. job_title = Ocupacion o cargo del trabajador 
5. salary = Salario recibido por el trabajador 
6. salary_currency = tipo de moneda del Salario 
7. salary_in_usd = Salario en dolares

9. remote_ratio = % de trabajo remoto 
10. company_location = Ubicacion de la compañia 
11. company_size = Tamaño da la compañia 

## Transformación de datos

Procedemos a transformar los datos 

```python
# Transformamos los datos de atributo experience_level
df['experience_level'] = df['experience_level'].replace(
    {'EN': 'Junior', 'MI': 'Intermedio', 'SE': 'Senior', 'EX': 'Experto'})
```

```python
# Transformamos los datos de atributo employment_type
df['employment_type'] = df['employment_type'].replace(
    {'FT': 'Fulltime', 'PT': 'Partime', 'FL': 'Freelance','CT': 'Contratado'})
```

```python
# Transformamos los datos de atributo company_size y generamosla categorizacion
df['company_size'] = df['company_size'].replace(
    {'S': 'Small', 'M': 'Medium', 'L': 'Grande'}).astype('category')
```

```python
# Incorporamos una nueva variable derivada de remote_ratio como employment_remote
df['employment_remote'] = df['remote_ratio'].replace(
    {100: 'Remote',50: 'Hibrido',0: 'Presencial' }).astype('category')
```

```python
df.sample(10)
```

Las transformaciones y agregación de datos que realizamos buscan facilitar la lectura para todos los interesados.  

## Análisis de datos

Visualizamos la informacion del dataset para comprender las relaciones de los atributos de manera rápida y realizar comparaciones e inferencias.

```python
# Primero veremos usando una matriz de correlacion de las variables numericas su dependencia entre ellas
sns.set_theme()
correlation_matrix = df.corr().round(2)
sns.set_theme(rc={'figure.figsize':(10, 8)})
sns.heatmap(data=correlation_matrix, annot=True)
plt.show()
```

Del resultado podemos inferir que el Salary_in_usd (salario en dolares) guarda una leve relación positiva con remote_ratio (trabajo remoto). Utilizaremos en su mayoria para el análisis comparativo el salario en dolares dado que permite una comparación uniforme entre los distintos paises.   
Vemos en el gráfico que Salary presenta correlación casi nula, será conveniente usarlo para análizar dentro del mismo grupo de pais mas no entre todos los paises.

```python
# Representamos la distribucion de los valores de los Salarios en dolares.
custom_param = {'figure.figsize': (10, 6)}
sns.set_theme(rc=custom_param)
sns.histplot(df['salary_in_usd'], bins=30, kde = True)
plt.show()
```

```python
df['salary_in_usd'].median()
```

Observamos que la distribucion de los salarios en dolares presenta un sesgo hacia la izquierda lo que nos indica que la mediana (USD 101.570) es más representativa que la media (USD 112297.869852) como valor del atributo.

```python
# Evaluaremos los salary_in_usd medios por pais, definido en un objeto contenedor - msl (median _salary_locations)
msl = df.groupby('company_location')['salary_in_usd'].median()
#msl
plt.figure(figsize=(12, 6))
sns.set(font_scale = 0.8)
sns.barplot(x = msl.index, y = msl.values)
```

```python
# Evaluaremos los salary_in_usd medios por tamaño de la empresa. Obj contenedor - mss (median _salary_size)
mss = df.groupby('company_size')['salary_in_usd'].median()
mss
```

Podemos visualizar que la mediana salarial según tamaño de la empresa es mayor en las de mediano porte, considerando los tres periodos del estudio.

```python
# Evaluaremos los salary_in_usd medios por tipo de contrato con la empresa. Obj contenedor - msc (median _salary_contract)
msc = df.groupby('employment_type')['salary_in_usd'].median()
msc
```

Podemos visualizar que la mediana salarial según tipo de empleo del trabajador es mayor para los contratados, considerando los tres periodos del estudio. 

```python
# Evaluaremos los salary_in_usd medios por forma de trabajo con la empresa. Obj contenedor - msr (median _salary_remote)
msr = df.groupby('employment_remote')['salary_in_usd'].median()
msr
```

Podemos visualizar que la mediana salarial según la forma de trabajo es mayor para los trabajadores en modalidad remoto, considerando los tres periodos del estudio. 

```python
# Agrupacion por localizacion, tamaño de la empresa, tipo de contrato y modalidad para mediana, minimos y maximos salariales.
dfm = df.groupby(['work_year','company_location','company_size','employment_type','employment_remote']).agg(
    {'salary_in_usd': ['count','median','max','min']}).reset_index()
dfm.sample(10)
```

```python
dfm.shape
```

```python
dfm = dfm.dropna()
dfm.sample(10)
```

```python
dfm.shape
```

Generamos un nuevo dataset con los atributos agrupados según la locación y tamaño de las empresas, ademas de tipo empleo y tipo contrato segmentados por año para obtener las medianas salariales, los minimos y maximos para cada grupo de categoria.
Con esto podemos visualizar las mismas agrupaciones realizadas anteriormente pero con mayor amplitud de caracteristicas.

## Visualizaciones

Con las visualizaciones logramos ver de manera resumida los datos de una forma más práctica y poder inferir conclusiones de las obsevaciones. 


```python
# Graficamos la relación entre el ingreso (salary_in_usd) y el trabajo remoto evaluando el tipo de trabajo
sns.relplot(x = 'employment_remote',
            y = 'salary_in_usd',
            data= df,
            hue='employment_type',
            col='employment_type',
            col_wrap=2)
plt.show()
```

Observamos una mayor distribucion de trabajadores full-time en las tres forma de trabajo remoto, hibrido y presencial (remote-ratio - 0 = presencial / 50 = hibrido / 100 = remoto) que en el resto de tipo de contrataciones.   
Diferente es para las otras formas de contrato que reflejan poca distribucion de datos; los trabajadores part-time(PT) solo tiene trabajadores de forma hibrida y remota; los freelancer(FL) tienen presencia en las tres formas de trabajo y los contratados(CT) lo hacen exclusivamente de forma remota. 

```python
# Queremos ver la cantidad de trabajadores segun su tipo de empleo y su nivel de experiencia 
plt.figure(figsize=(10, 6))
sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True)
sns.countplot(x = 'experience_level',
            hue='employment_type',
            data=df)
plt.xlabel('Nivel de Experiencia')
plt.title('Nivel de Experiencia segun tipo de empleo')
plt.show()
```

Corroboramos con esto que la mayor distribución de trabajadores está fulltime en su relación laboral, y al mismo tiempo el mayor numero de empleados tiene la categoria de Senior seguido de Intermedio dentro de la misma categoria de fulltime.

```python
# Evaluamos la dispersion de los salarios por año en función del tamaño de la empresa
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.set(font_scale=1.5)
sns.swarmplot(x='work_year',
            y='salary_in_usd',
            data=df,
            hue='company_size')
plt.xlabel('Años')
plt.ylabel('Salario USD')
plt.legend(title='Tamaño de la empresa', fontsize=10)
plt.show()
```

Se observa un crecimiento significativo de trabajadores año a año, la participación de trabajadores contratados en empresas de tamaña medio crece de 2020 a 2022 más, que en las empresas pequeñas y grandes.
